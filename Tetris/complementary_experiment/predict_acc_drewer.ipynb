```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
x = [1,2,3,4,5,6]
y_cp = [10.41,15.3,19.24,21.71,22.6,23.92]

# y_ca = [8.6,10.9,12.63,12.6,13.5,13.9]
y_ca = [8.6,10.9,12.63,12.6,13.5,13.9]
# y_mp = [1.11,1.25,1.37,1.8,2.08,2.06]
y_mp = [2.11,2.25,2.37,2.8,3.08,3.06]
# y_ma = [0.92,1.05,1.28,1.5,1.58,2.9]
y_ma = [0.92,1.05,1.28,1.5,1.58,1.9]

# y_cp=[9.8,9.81,9.8,9.75,9.77,10.21]
y_cp=[6.81,6.77,6.76,5.97,6.77,9.15]
# y_ca=[6.81,6.77,6.76,5.97,6.77,9.15]
y_ca=[9.8,9.81,9.8,9.75,9.77,10.21]
# y_mp=[2.89,2.99,3.1,3.97,3.32,1.95]
y_mp=[2.97,2.99,3.06,2.74,3.19,2.2]
# y_ma=[2.97,2.99,3.06,2.74,3.19,2.2]
y_ma=[3.89,3.99,4.1,4.97,4.32,2.95]
```

```python
fig, ax = plt.subplots(figsize=(12,9))
ax1 = ax
ax.set_axisbelow(True)

ax1.grid(axis = 'y')
ax1.set_xlabel('Prediction window size', family = 'Arial', fontsize = 46)
ax1.set_ylabel('MAPE of CPU (%)', family = 'Arial', fontsize = 46)
ax1.yaxis.label.set_color('#0e87cc')
ax1.set_ylim(0, 14)
ax1.plot(x, y_cp, label = 'Periodic CPU', color = '#0e87cc', linewidth = 4.0, marker='$\\bigodot$', ms=18)
ax1.plot(x, y_ca, label = 'Aperiodic CPU', color = '#0e87cc', linewidth = 4.0, marker='$\\bigodot$', ms=18, ls='--')

ax2 = ax1.twinx()
ax2.set_ylabel('MAPE of memory(%)', family = 'Arial', fontsize = 46)
ax2.yaxis.label.set_color('#db5856')
ax2.set_ylim(0, 14)
ax2.plot(x, y_mp, label = 'Periodic memory', color = '#db5856', linewidth = 4.0, marker='$\\bigtriangleup$', ms=18)
ax2.plot(x, y_ma, label = 'Aperiodic memory', color = '#db5856', linewidth = 4.0, marker='$\\bigtriangleup$', ms=18, ls='--')

ax1.set_xticks(np.array([1,2,3,4,5,6]))
ax1.tick_params(width = 4, length = 8, labelsize = 40, colors = '#0e87cc')
ax1.tick_params(axis = 'x', colors = 'black')
ax2.tick_params(width = 4, length = 8, labelsize = 40, colors = '#db5856')
bwith = 4
ax1.spines['top'].set_linewidth(bwith)
ax1.spines['bottom'].set_linewidth(bwith)
ax1.spines['left'].set_linewidth(bwith)
ax1.spines['right'].set_linewidth(bwith)
ax1.spines['left'].set_color('#0e87cc')
ax1.spines['right'].set_color('#db5856')
ax2.spines['left'].set_color('#0e87cc')
ax2.spines['right'].set_color('#db5856')

legend_elements = [
    Line2D([], [], marker='$\\bigodot$', color='#0e87cc', label='Peri. CPU', ls='-', linewidth = 3.0, ms=16),
    Line2D([], [], marker='$\\bigodot$', color='#0e87cc', label='Aperi. CPU', ls='--', linewidth = 3.0, ms=16),
    Line2D([], [], marker='$\\bigtriangleup$', color='#db5856', label='Peri. mem', ls='-', linewidth = 3.0, ms=16),
    Line2D([], [], marker='$\\bigtriangleup$', color='#db5856', label='Aperi. mem', ls='--', linewidth = 3.0, ms=16)
    ]
ax1.legend(handles=legend_elements, loc='center', bbox_to_anchor=(0.5, 0.87), ncol=2, columnspacing = 0.5, prop={'family': 'Arial', 'size': 37})
plt.savefig('./ARIMA.pdf', bbox_inches = 'tight', transparent = True)
plt.show()
```
